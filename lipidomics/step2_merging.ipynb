```python
import xlwings as xw
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.pyplot import figure
import numpy as np
import statistics

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("out_TommyCNL2025CLiverPos.csv")
```

```python
df = df.sort_values("Grade").sort_values("Adduct").sort_values("TopRT").sort_values("LipidMolec").sort_values("LipidGroupKey").drop(columns=["Unnamed: 0"])
```

```python
count = 0
start = 0
end = 0
tot = len(df)
can = list(df.columns)
for i in range(1, tot):
    
    if (df.iloc[i]["LipidMolec"] == df.iloc[i-1]["LipidMolec"]):
        continue
        
    else:
        end = i-1
        insert = list(df.iloc[i-1])
        insert[0] = "merged"
        insert[7] = np.mean(list(df["TopRT"])[start:end+1])
        for e in range(8, len(can)):
            insert[e] = np.sum(list(df[can[e]])[start:end+1])
        count -= 1
        df.loc[count] = insert
        start = i

insert = list(df.iloc[tot-1])
insert[0] = "merged"
insert[7] = np.mean(list(df["TopRT"])[start:tot])
for e in range(8, len(can)):
    insert[e] = np.sum(list(df[can[e]])[start:tot])
count -= 1
df.loc[count] = insert
        
df.index = df.index + abs(count)
df = df.sort_index()
df = df.sort_values("Grade").sort_values("Adduct").sort_values("TopRT").sort_values("LipidMolec").sort_values("LipidGroupKey").drop(columns=["Grade"])
df = df.loc[df['index'] == "merged"]
df.to_csv("out2_TommyCNL2025CLiverPos.csv")
```
