```python
def peakAligment(baseDf, inDf, n):
    count = 0
    tot = len(baseDf)
    canA = list(baseDf["LipidGroupKey"])
    canB = list(baseDf["LipidMolec"])
    for i in range(len(inDf)):
        if (inDf.iloc[i]["LipidGroupKey"] not in canA or inDf.iloc[i]["LipidMolec"] not in canB):
            insert = list(inDf.iloc[i])
            temp = n
            while temp > 0:
                insert.insert(-1, float(0))
                temp -= 1
            temp = 12 - n
            while temp > 0:
                insert.append(float(0))
                temp -= 1
            count -= 1
            #print(insert)
            baseDf.loc[count] = insert
        else:
            for j in range(tot):

                if (inDf.iloc[i]["LipidGroupKey"] == baseDf.iloc[j]["LipidGroupKey"]) and (inDf.iloc[i]["LipidMolec"] == baseDf.iloc[j]["LipidMolec"]) and (baseDf[baseDf.columns[8+n]][j] == float(0)) and (inDf.iloc[i]["TopRT"] > baseDf.iloc[j]["TopRT"]-0.05) and (inDf.iloc[i]["TopRT"] < baseDf.iloc[j]["TopRT"]+0.05):
                    baseDf[baseDf.columns[8+n]][j] = inDf["Area"][i]
                    #print(baseDf)
                    break

                elif (inDf.iloc[i]["LipidGroupKey"] == baseDf.iloc[j]["LipidGroupKey"]) and (inDf.iloc[i]["LipidMolec"] == baseDf.iloc[j]["LipidMolec"]) and ( (inDf.iloc[i]["TopRT"] < baseDf.iloc[j]["TopRT"]-0.05) or (inDf.iloc[i]["TopRT"] > baseDf.iloc[j]["TopRT"]+0.05) ):
                    insert = list(inDf.iloc[i])
                    temp = n
                    while temp > 0:
                        insert.insert(-1, float(0))
                        temp -= 1
                    temp = 12 - n
                    while temp > 0:
                        insert.append(float(0))
                        temp -= 1
                    count -= 1
                    #print("insert")
                    baseDf.loc[count] = insert
                    break

    baseDf.index = baseDf.index + abs(count)
    baseDf = baseDf.sort_index()
    baseDf.to_csv("out_TommyCNL2025CLiverPos.csv")
    return baseDf
```

```python
baseDf = peakAligment(baseDf=dfN1, inDf=dfN2, n=1)
list_df = [dfN4, dfN5, dfN7, dfT1, dfT2, dfT3, dfT4, dfT5, dfT7, df1QC, df2QC]
n = 1
for df in list_df:
    n += 1
    print(n)
    baseDf = peakAligment(baseDf=baseDf, inDf=df, n=n)
```
